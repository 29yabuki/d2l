_Regression_ problems aims to predict a quantitative value.

Common examples of regression are: predicting prices, predicting length of stay, forecasting demand, and many more.

Not all prediction problems fall under classical regression. Classification problems aims to predicts a quantitative value instead.

A dataset is called _training dataset_ or _training set_ in machine learning.

Each row of a dataset is an _example_ (also called _data point_, _instance_, or _sample_).

Each column is a distinct _feature_ (also called _covariates_) which are predictor measurements. More precisely, _features_ refer to a vector of $p$ predictor measurements.

A _label_ (also called _response_ or _target_) is the value that we're trying to predict.

Say we want to estimate the price of houses based on area and age.
- age and area are the features.
- price is the label.

## Basics

Linear regression is the simplest and most popular method for tackling regression problems.

Linear regressions follow a few key assumptions:
- The relationship between features $x$ and target $y$ is approximately linear such that the conditional mean $E[Y|X = x]$ can be expressed as a weighted sum of $x$.
- The target value may deviate from its expected value on the account of observation noise.
- The noise is assumed to follow a Gaussian (normal) distribution.

The proper notations are as follows:
- $n$ number of instances or examples.
- $\textbf{x}^{(i)}$ represents the $i^{\text{th}}$ instance.
- $\textbf{x}_j^{(i)}$ represents the $j^{\text{th}}$ coordinate.

### Model

Assuming linearity, the expected value of the target can be expressed as a weighted sum of the features, where:

$$
\text{price} = w_{\text{area}} \cdot \text{area} + w_{\text{age}} \cdot \text{age} + b
$$

where
- $w_{\text{area}}$ and $w_{\text{age}}$ are _weights_.
- $b$ is the _bias_ (also called _offset_ or _intercept_).

Weights determine the influence of a respective feature.

Bias determines the value of the estimate when all features are zero.

Bias is still necessary even when a house of zero are is impossible. Without it, we are limited to lines that only pass through the origin.

The recent equation is an _affine transformation_ of input features, characterized by
- a _linear transformation_ of features via the weighted sum, and
- a _translation_ via the added bias.

Given a dataset, our goal is to choose weights $\textbf{w}$ and the bias $b$ that on average, make our model's prediction fit the true prices as closely as possible.

Expressing models long-form like in the recent equation is common for disciplines that focus on datasets with only few features.

It is more convenient to express models with compact linear algebra notation in machine learning because it usually involves high-dimensional datasets.

A predictor $\hat y$, given inputs with $d$ features, can be expressed as

$$
\hat y = w_1 x_1 + \ldots + w_d x_d + b
$$

Accumulating all features into a vector $\textbf{x} \in \mathbb{R}^d$ and all weights into a vector $\textbf{w} \in \mathbb{R}^d$ gives us a compact expression for our model by using the dot product between $\textbf{w}$ and $\textbf{x}$,

$$
\hat y = \textbf{w}^{\top}\textbf{x} + b,
$$

where the vector $\textbf{x}$ correspdonds to the features of a single example.

It is often convenient to refer to features of a dataset of $n$ examples with the _design matrix_ $\textbf{X} \in \mathbb{R}^{n \times d}$.

The design matrix $\textbf{X}$ contains
- one row for every example or instance.
- one column for every feature.

The _predictions_ $\hat{\mathbf{y}} \in \mathbb{R}^n$ are computed using the matrix-vector product,

$$
{\hat{\mathbf{y}}} = \mathbf{X} \mathbf{w} + b,
$$

where $b$ is broadcasted along all entries.

The goal of linear regression given $d_{\text{train}}$ is to find both $\textbf{w}$ and $b$ such that when given $d_{\text{test}}$—the label of $d_{\text{test}}$ will be predicted with smallest error.

Before searching for the best model parameters $\textbf{w}$ and $b$, we need two more things:
- measure of quality of some given model, and
- a procedure for updating the model's quality improvement.

### Loss function

_Loss functions_ quantify the distance between the _real_ and _predicted_ values of the target. 

The loss function is usually a nonnegative negative where
- smaller values are better, and
- perfect predictions incur a loss of $0$.

The most common loss function for regression problems is the squared error.

Given a prediction $\hat y^{(i)}$ for an example $i$ and a corresponding true label $y^{(i)}$, the squared error is
$$
l^{(i)}(\textbf{w}, b) = \frac{1}{2}\left(\hat y^{(i)} - y^{(i)}\right)^2,
$$

where the constant $\frac{1}{2}$ makes no real difference, but is notationally convenient since it cancels out when we take the derivative of the loss.

Because the training dataset is given and it is out of our control—the _empirical error_ is only a function of the model parameters. 

Below shows a visualization of fitting a linear regression model to a one-dimensional data.

![](../../images/fit-linreg.svg)

Notice that large differences between estimates $\hat y^{(i)}$ and targets $y^{({i})}$ lead to larger contributions to the loss, due to the quadratic form.

The quadracity is a double-edged sword.
- It encourages the model to avoid large errors, but
- It can also lead to excessive sensitivity to anomalous data.

To measure the quality of a model on the entire dataset of $n$ examples, simply average the losses on $d_{\text{train}}$

$$
L(\textbf{w}, b) = \frac{1}{n} \sum_{i=1}^n l^{(i)}(\textbf{w}, b) = \frac{1}{n} \sum_{i=1} \frac{1}{2} \left( \textbf{w}^{\top} \textbf{x}^{(i)} + b - y^{(i)} \right)
$$

When training models, we search for parameters ($\textbf{w}^*$, $b^{*}$) that minimizes the total loss across all training examples

$$
\mathbf{w}^*, b^* = \operatorname*{argmin}_{\mathbf{w}, b}\  L(\mathbf{w}, b).
$$

### Analytic solution



Linear regression has a surprisngly simple optimization problem—find the optimal parameters analytically using a simple formula.



The bias $b$ is absorbed into the vector $\textbf{w}$ by adding a column of $1\textbf{s}$ to the design matrix $\textbf{X}$.

We minimize $\|\mathbf{y} - \mathbf{X}\mathbf{w}\|^2$ (difference between predicted and actual values) as the optimization objective.


When $\textbf{X}$ has a full rank, there is only one critical point and it is where there is minimum loss over the entire domain.

Taking the derivative of the loss with respect to $\textbf{w}$ and setting it to zero is

$$
\begin{aligned}
    \partial_{\mathbf{w}} \|\mathbf{y} - \mathbf{X}\mathbf{w}\|^2 =
    2 \mathbf{X}^\top (\mathbf{X} \mathbf{w} - \mathbf{y}) = 0
    \textrm{ and hence }
    \mathbf{X}^\top \mathbf{y} = \mathbf{X}^\top \mathbf{X} \mathbf{w}.
\end{aligned}
$$

Solving for $\textbf{w}$ gives us the optimal solution.

$$
\mathbf{w}^* = (\mathbf X^\top \mathbf X)^{-1}\mathbf X^\top \mathbf{y}
$$
- This only works when $\mathbf X^\top \mathbf X$ is invertible.

Most real-world machine learning problems are often too complex to incur an analytic solution, unlike linear regression. Get used to _iterative optimization_ to find good approximations.

The following are examples of iterative optimization techniques:
- gradient descent
- backpropagation
- stochastic methods

### Minibatch Stochastic Gradient Descent

_Gradient descent_ is an algorithm in optimizing nearly every deep learning model by iteratively reducing the error by updating the parameters in the direction that incrementally lowers the loss function.

The most naive application of gradient descent is the _batch gradient descent_ where it computes the loss derivative over the entire dataset before updating. It slow and inefficient, especially when training data is redundant, limiting its full benefit.

An other extreme application is the _stochastic gradient descent_ (SGD) algorithm—where you only consider a single example at a time and take update steps based on one observation at a time.

However, SGD has both computational and statistical drawbacks.
- Modern processors are faster at arithmetics operations than moving data from main memory to processor cache.
- Performing vector-vector multiplication is less efficient than a single large matrix-vector multiplication. Meaning, processing a sample at a time is slower compared to a full batch.
- Deep learning techniques such as batch normalization only work well with multiple observation at a time to compute stable statistics and reduce noise.

An intermediate approach called _minibatch stochastic gradient descent_ is used instead. Rather than taking a full batch or a single sample at a time, take a _minibatch_ of observations.

The size of the minibatch depends on many factors such as:
- amount of memory.
- number of accelerators.
- choice of layers.
- total dataset size.

Despite that, a number from 32 to 256 preferably a large power of $2$ are a good start.

The minibatch-based update step follows these steps:
1. Initialize the values of the model parameters, typically at random.
2. In each iteration $t$, we first randomly sample a minibatch $\mathcal{B}_t$ consisting of a fixed number $|\mathcal{B}|$ of training examples.
3. Compute for the gradient of the average loss on the minibatch with respect to the model parameters.
4. Multiply the gradient by a predetermined small value $\eta$ (called the _learning rate_). 
5. Subtract the resulting term from the current parameter values. This updates the parameters in the direction of the negative gradient.

All in all, the update is represented as:

$$
(\mathbf{w},b) \leftarrow (\mathbf{w},b) - \frac{\eta}{|\mathcal{B}|} \sum_{i \in \mathcal{B}_t} \partial_{(\mathbf{w},b)} l^{(i)}(\mathbf{w},b).
$$

For quadratic losses and affine transformations, it has the following closed-form expansion:

$$
\begin{split}\begin{aligned} \mathbf{w} & \leftarrow \mathbf{w} - \frac{\eta}{|\mathcal{B}|} \sum_{i \in \mathcal{B}_t} \partial_{\mathbf{w}} l^{(i)}(\mathbf{w}, b) && = \mathbf{w} - \frac{\eta}{|\mathcal{B}|} \sum_{i \in \mathcal{B}_t} \mathbf{x}^{(i)} \left(\mathbf{w}^\top \mathbf{x}^{(i)} + b - y^{(i)}\right)\\ b &\leftarrow b -  \frac{\eta}{|\mathcal{B}|} \sum_{i \in \mathcal{B}_t} \partial_b l^{(i)}(\mathbf{w}, b) &&  = b - \frac{\eta}{|\mathcal{B}|} \sum_{i \in \mathcal{B}_t} \left(\mathbf{w}^\top \mathbf{x}^{(i)} + b - y^{(i)}\right). \end{aligned}\end{split}
$$

Picking a minibatch $\mathcal{B}$ means we need to normalize by its size $\vert \mathcal{B} \vert$.

Frequently, $\vert \mathcal{B} \vert$ and $\eta$ are user-defined. These are called _hyperparameters_ where they are tunable parameters that are not updated in the training loop.

Hyperparameters are tuned automatically by a number of techniques such as Bayesian optimization.

Ultimately, the quality of the solution is typically assessed on a separate _validation dataset_.

After training some predetermined number of iterations (or until certain stopping criterion is met), we record the estimated model parameters as ${\hat{\textbf{w}}}$ and $\hat b$.

Even if our function is truly linear and noiseless, the randomness in picking $\mathcal{B}$ prevents the estimated parameters from being exact minimizers of the loss and makes them non-deterministic. As a result, while the algorithm heads toward the optimal solution, it may never reach it exactly.

As previously mentioned, in linear regression there is a global minima when $\textbf{X}$ is full rank. However, loss surfaces for deep learning contains many saddle points and minima. Therefore, the goal is not to find an exact minimizer. Rather, find a set of pararmeters that lead to accurate predictions (and thus low loss).

In practice, deep learning practitioners seldom struggle to find parameters that minizes the loss on _training sets_.

The more formidable task is to find parameters that lead to accurate predictions on previously unseen data. This is called _generalization_.

### Predictions

Given the model,
$$
\hat{\mathbf{w}}^\top \mathbf{x} + \hat{b},$$ 

we are able to make predictions for a new example.

Deep learning uses the term _inference_ for the prediction phase, it is bit of a misnomer. In statistics, _inference_ refers to estimating the underlying distributions or relationsh. This can cause confusion—stick with the term _prediction_.

## Vectorization for speed

In order to process whole minibatches of examples simultaneously, we need vectorize the calculations and leverage fast linear algebra libraries instead of writing for-loops.

```python
n = 10000
a = torch.ones(n)
b = torch.ones(n)
```

The first method below loop over the vectors with a for-loop.

```python
c = torch.zeros(n)
t = time.time()
for i in range(n):
    c[i] = a[i] + b[i]
print(f'{time.time() - t:.5f} sec')
```

The second (and faster) method is we rely only on the reloaded `+` operator.

```python
t = time.time()
d = a + b
print(f'{time.time() - t:.5f} sec')
```

Additionally, the second method reduces the potential for error and increases portability.

## The normal distribution and squared loss

A normal distribution with mean $\mu$ and variance $\sigma^2$ is 

$$
p(x) = \frac{1}{\sqrt{2 \pi \sigma^2}} \exp\left(-\frac{1}{2 \sigma^2} (x - \mu)^2\right).
$$

We define a function that computes the normal distribution as

```python
def normal(x, mu, sigma):
    p = 1 / math.sqrt(2 * math.pi * sigma**2)
    return p * np.exp(-0.5 * (x - mu)**2 / sigma**2)
```

We can also visualize the normal distributions.

```python
x = np.arange(-7, 7, 0.01)

# mean and std. deviation pairs
params = [(0, 1), (0, 2), (3, 1)]
d2l.plot(x, [normal(x, mu, sigma) for mu, sigma in params], 
        xlabel='x', ylabel='p(x)', figsize=(4.5, 2.5),
        legend=[f'mean {mu}, std {sigma}' for mu, sigma in params])
```

Remember changing the mean shifts the distribution along the $x$-axis. On the other hand, changing the variance spreads the distribution while lowering its peak.

A way to motivate linear regression with squared loss is by assuming that observations come from noisy measurements, where the noise $\epsilon$ follows the normal distribution $\mathcal{N}(0, \sigma^2)$
$$
y = \mathbf{w}^\top \mathbf{x} + b + \epsilon \textrm{ where } \epsilon \sim \mathcal{N}(0, \sigma^2).
$$

Therefore, the _likelihood_ for a particular $y$ for given $\textbf{x}$ is
$$
P(y \mid \mathbf{x}) = \frac{1}{\sqrt{2 \pi \sigma^2}} \exp\left(-\frac{1}{2 \sigma^2} (y - \mathbf{w}^\top \mathbf{x} - b)^2\right).
$$

The _likelihood_ function is
$$
P(\mathbf y \mid \mathbf X) = \prod_{i=1}^{n} p(y^{(i)} \mid \mathbf{x}^{(i)}),
$$

where the best values of $\mathbf{w}$ and $b$ maximizes the likelihood of the dataset.

Maximizing the product of many exponential functions is difficult. To simplify this, maximize the logarithm of the likelihood instead.

But for historical reasons, optimizations are expressed as minimization rather than maximization. Thus, we minimize the _negative log-likelihood_.
$$
-\log P(\mathbf y \mid \mathbf X) = \sum_{i=1}^n \frac{1}{2} \log(2 \pi \sigma^2) + \frac{1}{2 \sigma^2} \left(y^{(i)} - \mathbf{w}^\top \mathbf{x}^{(i)} - b\right)^2
$$

Minimizing mean squared error is equivalent to maximum likelihood estimation for a linear model with additive Gaussian noise for the reasons below.
- The first term is independent on $\textbf{w}$ or $b$ if $\sigma$ is fixed.
- The second term is equivalent to the squared error loss except for the factor $\frac{1}{\sigma^2}$.

## Linear regression as a neural network

Linear models lack the complexity to represent many complicated networks. Fortunately, neural networks are rich enough to contain linear models as networks—where every feature is represented by an input neuron, all of which are connected to the output.

![](../../images/singleneuron.svg)

The inputs are $x_1, \ldots, x_d$ where $d$ is the number of inputs or the _feature dimensionality_. 
All inputs are given. The network output is single computed neuron $o_1$.

## Biology

A biological neuron consists of
- dendrites (input terminals), 
- the nucleus (CPU), 
- the axon (output wire),
- and the axon terminals (output terminals).

![](../../images/neuron.svg)

They enable connections to other neurons via synapses.

Information $x_1$ from other neurons is received in the dendrites. Additionally, that information $x_io$ is weighted by synaptic weights $x_i$ determining activation or inhibition with the product $x_i w_i$. 

The weighted inputs arriving from multiple sources are aggregated in the nucleus as a weighted sum $y = \sum_i x_i w_i + b$, where it is subject to some nonlinear postprocessing via $\sigma(y)$. The axon then transmits this information to terminals, reaching another neuron's dendrites or an actuator such as a muscle.
